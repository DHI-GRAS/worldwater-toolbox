```python
from openeo.udf import XarrayDataCube
from openeo.udf.debug import inspect
import openeo

connection = openeo.connect("openeo-dev.vito.be").authenticate_oidc()

cube = connection.load_collection(
  "SENTINEL1_GRD",
  spatial_extent={"west": 16.06, "south": 48.06, "east": 16.65, "north": 48.35},
  temporal_extent=["2017-03-01", "2017-04-01"],
  bands=["VV", "VH"])

udf_code = """

def apply_datacube(cube: XarrayDataCube, context: dict) -> XarrayDataCube:
    inspect(message="Hello UDF logging")
    inspect(data=[1,2,3], message="hello UDF logging with data")
    return cube
"""

udf  = openeo.UDF(code=udf_code, runtime="Python", data={"from_parameter": "x"})
out_cube = cube.apply(process=udf )

my_job  = cube.create_job(title="test")
results = my_job.start_and_wait().get_results()
```
